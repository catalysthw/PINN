```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
torch.manual_seed(42)
```

```python
# ==============================================================================
# Global Parameters (Placeholders - Define these based on your specific problem)
# ==============================================================================
# Define constants
X_MIN, X_MAX = 0.0, 1.0  # Spatial domain [0, L]
T_MIN, T_MAX = 0.0, 1.0  # Time domain [0, T]
L = X_MAX - X_MIN
n = 1  # Mode for initial condition (e.g., sin(n*pi*x/L))

# Placeholder for u_exact function (necessary for evaluation)
def u_exact(x, t, alpha):
    """ Calculates the analytical solution for the 1D Heat Equation."""
    global L, n
    # u(x,t) = exp(-n^2 * pi^2 * alpha * t / L^2) * sin(n * pi * x / L)
    # alpha here is a scalar value (not tensor expanded)
    return np.exp(-n**2 * np.pi**2 * alpha * t / L**2) * np.sin(n * np.pi * x / L)
```

```python
# ==============================================================================
# 1. MODEL DEFINITION (PINN_Net Class)
# ==============================================================================

class PINN_Net(nn.Module):
    """
    Neural Network for the Meta-PINN model.
    It takes (x, t, alpha) as input (3 features).
    """
    def __init__(self, input_features=3, hidden_layers=4, hidden_units=50):
        super(PINN_Net, self).__init__()
        
        # Input layer (3: x, t, alpha)
        layers = [nn.Linear(input_features, hidden_units), nn.Tanh()]
        
        # Hidden layers
        for _ in range(hidden_layers):
            layers.extend([nn.Linear(hidden_units, hidden_units), nn.Tanh()])
            
        # Output layer (1: u(x, t, alpha))
        layers.append(nn.Linear(hidden_units, 1))
        
        self.fcs = nn.Sequential(*layers)

    def forward(self, x):
        """
        Forward pass. x is expected to be a tensor of shape [Batch Size, 3] (x, t, alpha).
        """
        return self.fcs(x)


# ==============================================================================
# 2. CORE PINN FUNCTIONS
# ==============================================================================

# Sampling points (Collocation points-interior, initial/boundary condition points)
def sample_points(N_f=5000, N_i=200, N_b=200):
    """
    Generates static collocation points for the PDE (interior), Initial Conditions (IC), 
    and Boundary Conditions (BC).
    """
    # 1. Collocation points (interior) - x and t are sampled uniformly (Random sampling)
    x_f = torch.FloatTensor(N_f, 1).uniform_(X_MIN, X_MAX).to(device)
    t_f = torch.FloatTensor(N_f, 1).uniform_(T_MIN, T_MAX).to(device)

    # 2. Initial condition points (t = T_MIN) - x is sampled evenly (Linspace)
    x_i = torch.linspace(X_MIN, X_MAX, N_i).view(-1, 1).to(device)
    t_i = torch.zeros_like(x_i).to(device)
    # Exact IC value: u(x, 0) = sin(n * pi * x / L)
    u_i = torch.sin(n * np.pi * x_i / L).to(device) 
    
    # 3. Boundary conditions points (x = X_MIN/X_MAX) - t is sampled evenly (Linspace)
    t_b = torch.linspace(T_MIN, T_MAX, N_b).view(-1, 1).to(device)
    x_b0 = X_MIN * torch.ones_like(t_b).to(device) # Left boundary (x=0)
    x_b1 = X_MAX * torch.ones_like(t_b).to(device) # Right boundary (x=L)

    return (x_f, t_f), (x_i, t_i, u_i), (x_b0, x_b1, t_b)


# Note: pde_residual function is assumed to be defined as follows:
# def pde_residual(model, x, t, alpha)
def pde_residual(model, x, t, alpha):
    """
    Calculates the residual of the 1D Heat Equation: u_t - alpha * u_xx.
    This function handles the input where alpha is a scalar tensor for the PDE term.
    """
    # Enable gradient tracking for the input coordinates (x and t)
    x.requires_grad_(True)
    t.requires_grad_(True)
    
    # Expand the scalar alpha tensor to match the batch size of x for model input
    alpha_input = alpha.expand_as(x) 
    
    # Forward pass: get the predicted solution u(x, t, alpha)
    u = model(torch.cat([x, t, alpha_input], dim=1))
    
    # Calculate derivatives using Autograd (create_graph=True for the subsequent loss backpropagation)
    # 1st Derivative w.r.t time (u_t)
    u_t = torch.autograd.grad(u, t, torch.ones_like(u), create_graph=True)[0]
    # 1st Derivative w.r.t space (u_x)
    u_x = torch.autograd.grad(u, x, torch.ones_like(u), create_graph=True)[0]
    # 2nd Derivative w.r.t space (u_xx = derivative of u_x w.r.t x)
    u_xx = torch.autograd.grad(u_x, x, torch.ones_like(u_x), create_graph=True)[0]

    # Return the PDE residual (f = u_t - alpha * u_xx)
    # The scalar tensor alpha is used for the physics term.
    return u_t - alpha * u_xx


def pinn_loss(model, alpha_tensor, pts):
    """
    Calculates the total PINN loss (L_f + L_i + L_b) for a specific alpha value.
    The model takes (x, t, alpha) as a 3D input.
    """
    (x_f, t_f), (x_i, t_i, u_i), (x_b0, x_b1, t_b) = pts
    
    # Expand scalar alpha tensor to match the shape of coordinates for model input concatenation
    alpha_f = alpha_tensor.expand_as(x_f)
    alpha_i = alpha_tensor.expand_as(x_i)
    alpha_b = alpha_tensor.expand_as(x_b0)
    
    # 1. PDE Loss (L_f): Mean Squared Error of the PDE residual
    loss_f = torch.mean(pde_residual(model, x_f, t_f, alpha_f)**2) 
    
    # 2. Initial Condition Loss (L_i): MSE of (Predicted IC - Exact IC)
    loss_i = torch.mean((model(torch.cat([x_i, t_i, alpha_i], dim=1)) - u_i)**2)
    
    # 3. Boundary Condition Loss (L_b): MSE of Predicted BC (assuming u=0 on boundaries)
    loss_b = torch.mean(model(torch.cat([x_b0, t_b, alpha_b], dim=1))**2 +
                        model(torch.cat([x_b1, t_b, alpha_b], dim=1))**2)
                        
    return loss_f + loss_i + loss_b


# ==============================================================================
# 3. META-LEARNING FUNCTIONS (Reptile)
# ==============================================================================

def meta_train(meta_model, alpha_list, meta_steps=300, inner_steps=3, meta_lr=1e-3, inner_lr=1e-2):
    """
    Performs Reptile-style Meta-Training (Outer Loop).
    Optimizes the initial weights (meta_model, phi) for fast adaptation.
    """
    # Outer Loop: Meta-Learning Steps
    for step in range(1, meta_steps + 1):
        
        # 1. Store the initial meta weights (phi) before starting the tasks.
        # This is the starting point for the Inner Loop.
        meta_weights_phi = {}
        for k, v in meta_model.state_dict().items():
            # Deep copy the tensor data to protect phi from inner loop updates.
            meta_weights_phi[k] = v.clone()
        
        # 2. Loop through each Task (alpha_i) in the Meta-Training set
        for alpha in alpha_list:
            alpha_tensor = torch.tensor([[alpha]], device=device)
            
            # --- Inner Loop Setup ---
            # Create a temporary model (W) initialized with meta weights (phi)
            temp_model = PINN_Net().to(device) 
            temp_model.load_state_dict(meta_weights_phi) # Load the initial weights (phi)

            # Use SGD for inner loop optimization
            optimizer = torch.optim.SGD(temp_model.parameters(), lr=inner_lr)
            pts = sample_points() 

            # --- Inner Loop Training (k steps) ---
            # Train the temporary model on the specific task (alpha) to get W (post-update weights).
            # The number of steps, 'k' (inner_steps), is intentionally small for fast adaptation.
            for _ in range(inner_steps):
                # 1. Forward Pass: Calculate the Physics-Informed Loss for the current task (alpha).
                loss = pinn_loss(temp_model, alpha_tensor, pts) 
                # 2. Reset Gradients: Clear accumulated gradients.
                optimizer.zero_grad() 
                # 3. Backward Pass: Compute gradients (dLoss/dW).
                loss.backward() 
                # 4. Optimizer Step: Update temp_model's weights (W) using inner_lr.
                optimizer.step()
            
            # 3. Meta Update (Outer Loop: Reptile Step)
            # Update phi in the direction of the inner update: phi <- phi + epsilon * (W - phi)
            with torch.no_grad():
                for name, param in meta_model.named_parameters():
                    # (W - phi): Vector representing the direction of adaptation (the "good" direction learned).
                    # param.data is the current meta weight (phi_old).
                    param.data += meta_lr * (temp_model.state_dict()[name] - meta_weights_phi[name])

        # Logging
        if step % 50 == 0:
            print(f"[Meta Step {step}] Done.")
    return meta_model


def fine_tune_from_meta(meta_model, alpha_new, steps=300, lr=1e-3):
    """
    Fine-tunes the meta-learned initial model (meta_model) on a new, specific task (alpha_new).
    """
    # Initialize model with meta-learned weights (best initial guess)
    model = PINN_Net().to(device)
    model.load_state_dict(meta_model.state_dict())
    
    # Use Adam optimizer for fine-tuning
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    alpha_tensor = torch.tensor([[alpha_new]], device=device)
    
    # Fine-tuning Loop
    for step in range(1, steps + 1):
        pts = sample_points() 
        loss = pinn_loss(model, alpha_tensor, pts)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        if step % 100 == 0 or step == 1:
            print(f"FT Step {step}/{steps}, Loss = {loss.item():.2e}")
            
    return model

# ==============================================================================
# 4. EVALUATION AND VISUALIZATION FUNCTIONS
# ==============================================================================

def plot_results(u_pred, u_true, X, T, alpha, mse, title="Meta-PINN"):
    """
    Visualizes the PINN prediction and the exact solution using contour plots.
    """
    plt.figure(figsize=(12, 4))
    
    # Plot 1: PINN Prediction
    plt.subplot(1, 2, 1)
    plt.contourf(X, T, u_pred, 100, cmap="plasma")
    plt.title(f"Meta-PINN Prediction (α={alpha})")
    plt.xlabel("x"); plt.ylabel("t")
    plt.colorbar(label='u(x,t)')
    
    # Plot 2: Exact Solution
    plt.subplot(1, 2, 2)
    plt.contourf(X, T, u_true, 100, cmap="plasma")
    plt.title("Exact Solution")
    plt.xlabel("x"); plt.ylabel("t")
    plt.colorbar(label='u(x,t)')
    
    # Overall Title and Layout
    plt.suptitle(f"{title} | MSE = {mse:.2e}", fontsize=14)
    plt.tight_layout()
    plt.show()


def evaluate(model, alpha, title="Meta-PINN", show=True):
    """
    Evaluates the model's prediction, calculates MSE, and optionally visualizes the results.
    """
    # 1. Prepare Test Grid (100x100 points)
    x_vals = np.linspace(X_MIN, X_MAX, 100)
    t_vals = np.linspace(T_MIN, T_MAX, 100)
    X, T = np.meshgrid(x_vals, t_vals)
    
    # Convert and prepare inputs
    x_test = torch.tensor(X.flatten()[:, None], dtype=torch.float32).to(device)
    t_test = torch.tensor(T.flatten()[:, None], dtype=torch.float32).to(device)
    # Expanded alpha input for the 3D model input (x, t, alpha)
    alpha_test = torch.full_like(x_test, alpha).to(device)

    # 2. Model Prediction
    model.eval()
    with torch.no_grad():
        u_pred = model(torch.cat([x_test, t_test, alpha_test], dim=1)).cpu().numpy().reshape(100, 100)
        
    # 3. Reference Solution and Error
    u_true = u_exact(X, T, alpha)
    mse = np.mean((u_pred - u_true)**2)
    
    # 4. Visualization (Conditional Call)
    if show:
        plot_results(u_pred, u_true, X, T, alpha, mse, title)
        
    return mse


# ==============================================================================
# 5. TEST EXECUTION BLOCK
# ==============================================================================
# 1. DEFINE TRAINING TASKS (alpha_training_set)
print("1. Defining Alpha Training Set (Tasks)...")
alpha_train_list = [0.1, 0.2, 0.3, 0.4, 0.6, 0.8, 0.9]

# meta_model = meta_train(initial_net, alpha_train_list, ...)
initial_meta_model = PINN_Net().to(device)
print("2. Initializing Model and Hyperparameters...")

# Define Meta-Training Hyperparameters
META_STEPS = 1000  
INNER_STEPS = 5
META_LR = 1e-3
INNER_LR = 1e-2

meta_model = meta_train(
    meta_model=initial_meta_model, 
    alpha_list=alpha_train_list, 
    meta_steps=META_STEPS, 
    inner_steps=INNER_STEPS, 
    meta_lr=META_LR, 
    inner_lr=INNER_LR
)

print("Meta-Training Successfully Finished. 'meta_model' is now ready for evaluation.")


# Meta-learning test 
print("\n4. Starting Meta-PINN Evaluation on Unseen Tasks...")

# alpha test set
alphas_to_test = [0.35, 0.45, 0.55, 0.70, 0.01]
mse_raw = []        
mse_finetuned = []  

for alpha in alphas_to_test:
    print("-" * 30)
    print(f"--- Evaluating Task: α = {alpha} ---")
    
    # 1. Raw Evaluation
    print("  -> 1. Evaluating Raw Meta-PINN (no fine-tuning)...")
    mse1 = evaluate(model=meta_model, alpha=alpha, title=f"Meta-PINN (α={alpha}, Raw)", show=False)
    print(f"     Raw MSE: {mse1:.2e}")
    mse_raw.append(mse1)

    # 2. Fine-Tuning
    print("  -> 2. Fine-tuning Meta-PINN...")
    ft_model = fine_tune_from_meta(meta_model=meta_model, alpha_new=alpha, steps=1000, lr=1e-3)

    # 3. Fine-tuned Evaluation
    print("  -> 3. Evaluating Fine-tuned Meta-PINN...")
    mse2 = evaluate(model=ft_model, alpha=alpha, title=f"Fine-tuned Meta-PINN (α={alpha})", show=True)
    print(f"     Fine-tuned MSE: {mse2:.2e}")
    mse_finetuned.append(mse2)

print("\n" + "=" * 50)
print("Evaluation Summary (MSE)")
print(f"Raw Meta-PINN MSEs: {np.array(mse_raw)}")
print(f"Fine-tuned MSEs:    {np.array(mse_finetuned)}")
print("=" * 50)
```
